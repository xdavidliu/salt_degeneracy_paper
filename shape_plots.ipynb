```python
%matplotlib inline
import numpy as np
import pylab as pl
```

```python
x = np.linspace(0, 2*np.pi)
p1, = pl.plot(x, np.sin(x), 'r-')
p2, = pl.plot(x, np.cos(x), 'b-')
pl.xlabel('foobar')
pl.ylabel('yahoo')
pl.title('awesome plot')
pl.legend([p1, p2], ['sine', 'cosine'], loc='best')
pl.show()
```

```python
data = np.random.normal(5, 3, 1000)
bins = np.arange(-5, 16, 1)
pl.hist(data, bins, histtype='stepfilled')
pl.show()
```

```python
from matplotlib.patches import Rectangle
fig = pl.figure()
ax = pl.gca()
# axes labels make figure unsymmetric; turn them off.
# ax.get_xaxis().set_visible(False)
# ax.get_yaxis().set_visible(False)
# this doesn't seem to work, after savefig, still unsymmetric
ax.add_patch(Rectangle((-0.5, -0.5), 1, 1, facecolor='#94acb5'))
pl.axis('equal')
pl.axis('off')
pl.xlim(-0.6, 0.6)
pl.ylim(-0.6, 0.6)
pl.savefig('square.pdf')
# completely tight bounding box at
# /MediaBox [128 54 315 241]
# reasonably bordered bounding box at
# /MediaBox [88 14 355 281]
```
